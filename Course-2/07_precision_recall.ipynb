## Confusion Matrix


| Predicted / Actual | 1              | 0              |
|--------------------|----------------|----------------|
| 1                  | True Positive  | False Positive |
| 0                  | False Negative | True Negative  |


$ Precision \to what\ fraction\ of\ the\ predictions\ is\ actually\ true? \\
\\
 Precision = \frac{True\ Positives}{True\ Positives + False\ Positives} $

$ Recall \to what\ fraction\ of\ all\ the\ positives\ is\ correctly\ predicted? \\
\\
 Recall = \frac{True\ Positives}{True\ Positives + False\ Negatives} $

$Precision\ and\ Recall\ are\ metrics\ of\ measuring\ the\ accuracy\ of\ a\ learning\ algorithm.$
