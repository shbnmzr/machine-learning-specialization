```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.keras import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.activations import relu, linear
from tensorflow.python.keras.losses import SparseCategoricalCrossentropy
from sklearn.model_selection import train_test_split
```

```python
def import_data(path):
    data = pd.read_csv(path)
    return data
```

```python
data = import_data('./digit-recognizer/train.csv')
```

```python
def split_train_dev_test(data, train_split=0.7, dev_split=0.15, test_split=0.15):
    train_set, rest = train_test_split(data, test_size=dev_split+test_split, train_size=train_split, random_state=1)
    dev_set, test_set = train_test_split(rest, test_size=0.5, train_size=0.5, random_state=1)
    return train_set, dev_set, test_set
```

```python
train_set, dev_set, test_set = split_train_dev_test(data)
```

```python
def split_x_y(data, target):
    y = data[target]
    X = data.drop([target], axis=1)
    return np.array(X), np.array(y)
```

```python
print(f'Training Set: {len(train_set)}')
print(f'Cross Validation Set: {len(dev_set)}')
print(f'Test Set: {len(test_set)}')
```

```python
X_train, y_train = split_x_y(train_set, 'label')
X_dev, y_dev = split_x_y(dev_set, 'label')
X_test, y_test = split_x_y(test_set, 'label')
```

```python
def create_model(num_layers, units, activations):
    layers = []
    for i in range(num_layers):
        layers.append(Dense(units=units[i], activation=activations[i]))

    model = Sequential(layers)
    return model
```

```python
model_1 = create_model(4, [30, 25, 15, 10], ['relu', 'relu', 'relu', 'linear'])
```

```python
model_1.compile(optimizer='adam', loss=SparseCategoricalCrossentropy())
```

```python
model_1.fit(X_train, y_train, epochs=50)
```
