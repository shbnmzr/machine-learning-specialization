## Simple Neural Network with Python

```python
import numpy as np
import tensorflow as tf
```

## DataSet

```python
X = np.random.rand(200, 2) * 200
y = np.random.rand(200, 1)

for i in range(len(y)):
    y[i] = 1 if y[i] >= 0.5 else 0
```

```python
normalized = tf.keras.layers.Normalization(axis=-1)
normalized.adapt(X)

Xn = normalized(X)
```

```python
def sigmoid(z):
    g_z = 1 / (1 + np.exp(-z))
    return g_z
```

```python
def dense(a_in, W, b, g):
    z = np.matmul(a_in, W) + b
    a_out = g(z)
    return a_out
```

```python
def sequential(X, W1, b1, W2, b2):
    a1 = dense(X, W1, b1, sigmoid)
    a2 = dense(a1, W2, b2, sigmoid)
    f_x = a2
    return f_x
```

```python
W1_tmp = np.array( [[-8.93,  0.29 ], [-0.1,  -7.32 ]] )
b1_tmp = np.array( [[-9.82, -9.28 ]] )
W2_tmp = np.array( [[-31.18], [-27.59]] )
b2_tmp = np.array( [[15.41]] )
```

```python
def predict(X, W1, b1, W2, b2):
    m = X.shape[0]
    p = np.zeros((m, 1))
    for i in range(m):
        p[i, 0] = sequential(X[i], W1, b1, W2, b2)

    return p
```

```python
model = sequential(Xn, W1_tmp, b1_tmp, W2_tmp, b2_tmp)
```
