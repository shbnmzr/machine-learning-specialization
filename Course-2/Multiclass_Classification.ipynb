```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.python.keras import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.activations import linear, relu
from tensorflow.python.keras.losses import SparseCategoricalCrossentropy
from tensorflow.python.keras.regularizers import L2
```

```python
def import_data(path):
    data = pd.read_csv(path)
    return data
```

```python
def split_training_dev_test(data, training_size=0.8):
    training_set, dev_set = train_test_split(data, train_size=training_size, test_size=(1 - training_size), random_state=12)
    return training_set, dev_set
```

```python
def extract_target(data, target):
    X = data.drop([target], axis=1)
    y = data[target]
    return np.array(X), np.array(y)
```

```python
def layers_construction(num_layers, units, activations, lambda_=0.01):
    layers = []
    for i in range(num_layers):
        layers.append(Dense(units=units[i], activation=activations[i], kernel_regularizer=L2(lambda_)))
    return layers
```

```python
def model_construction(layers):
    model = Sequential(layers)
    return model
```

```python
def label_data(f_x):
    labels = np.array([np.argmax(i) for i in f_x])
    return labels
```

```python
def evaluate_model(yhat, y):
    score = [0 if yhat[i] == y[i] else 1 for i in range(len(y))]
    score = np.sum(score)
    score /= len(y)
    return score
```
