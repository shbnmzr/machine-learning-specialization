## Digit Recognition using Multiclass Classification

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.losses import SparseCategoricalCrossentropy
from tensorflow.python.keras import Sequential
from tensorflow.python.keras.optimizer_v1 import Adam
```

```python
data = pd.read_csv('./digit-recognizer/train.csv')
data.head()
```

```python
y_train = data['label']
X_train = data.drop('label', axis=1)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
model = Sequential([
    Dense(units=30, activation='relu'),
    Dense(units=25, activation='relu'),
    Dense(units=15, activation='relu'),
    Dense(units=10, activation='linear')
])
```

```python
model.compile(optimizer='adam', loss=SparseCategoricalCrossentropy(from_logits=True))
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
test_data = pd.read_csv('./digit-recognizer/test.csv')
test_data = np.array(test_data)
```

```python
logits = model(test_data)
```

```python
f_x = tf.nn.softmax(logits)
```

```python
label = np.array([np.argmax(i) for i in f_x])
```

```python
label
```

```python
label_df = pd.DataFrame(label)
```

```python
index = len(label_df)
index = [i + 1 for i in range(index)]
```

```python
label_df.index = index
label_df.index.names = ['ImageId']
label_df.columns = ['label']
```

```python
label_df.to_csv('digit-recognizer/submission.csv')
```
