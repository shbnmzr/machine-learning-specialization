## Digit Recognition using Multiclass Classification

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.losses import SparseCategoricalCrossentropy
from tensorflow.python.keras import Sequential
from sklearn.model_selection import train_test_split
from tensorflow.python.keras.regularizers import L2
```

```python
data = pd.read_csv('./digit-recognizer/train.csv')
data.head()
```

```python
training, dev = train_test_split(data, test_size=0.1, train_size=0.9)
```

```python
y_train, y_dev = training['label'], dev['label']
X_train, X_dev = training.drop('label', axis=1), dev.drop('label', axis=1)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
model = Sequential([
    Dense(units=65, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=60, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=55, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=50, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=45, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=40, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=35, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=30, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=25, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=20, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=15, activation='relu', kernel_regularizer=L2(0.00001)),
    Dense(units=10, activation='linear', kernel_regularizer=L2(0.00001))
])
```

```python
model.compile(optimizer='adam', loss=SparseCategoricalCrossentropy(from_logits=True))
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
X_dev = np.array(X_dev)
y_dev = np.array(y_dev)
logits_dev = model(X_dev)
f_x_dev = tf.nn.softmax(logits_dev)
yhat = np.array([np.argmax(i) for i in f_x_dev])
score = [0 if yhat[i] == y_dev[i] else 1 for i in range(len(y_dev))]
score = np.sum(score)
print(score / len(y_dev))
```

```python
test_data = pd.read_csv('./digit-recognizer/test.csv')
test_data = np.array(test_data)
```

```python
logits = model(test_data)
```

```python
f_x = tf.nn.softmax(logits)
```

```python
label = np.array([np.argmax(i) for i in f_x])
```

```python
label
```

```python
label_df = pd.DataFrame(label)
```

```python
index = len(label_df)
index = [i + 1 for i in range(index)]
```

```python
label_df.index = index
label_df.index.names = ['ImageId']
label_df.columns = ['label']
```

```python
label_df.to_csv('digit-recognizer/submission.csv')
```
