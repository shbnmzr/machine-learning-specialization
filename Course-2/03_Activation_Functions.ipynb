```python
import numpy as np
import matplotlib.pyplot as plt
```

## Activation Functions
### ReLu
$$ g\big(z\big) = max\big(0, z\big) $$

### Sigmoid
$$ g\big(z\big) = \frac{1}{1 + e^{-z}} $$

### Linear Regression
$$ g\big(z\big) = z $$

```python
def relu_(z):
    g = np.max((0., z))
    return g
```

```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
def linear(z):
    g = z
    return g
```

```python
X = np.arange(-10, 10, 0.5, )
y_relu = np.array([relu_(x) for x in X])
y_sigmoid = sigmoid(X)
y_linear = linear(X)

fig, ax = plt.subplots(2, 2, figsize=(16, 16))

ax[0, 0].plot(X, y_relu)
ax[0, 0].set_title('ReLU Activation Function')

ax[0, 1].plot(X, y_sigmoid)
ax[0, 1].set_title('Sigmoid Activation Function')

ax[1, 0].plot(X, y_linear)
ax[1, 0].set_title('Linear Activation Function')

plt.show()
```
