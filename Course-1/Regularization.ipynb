# Regularized Implementation of Linear Regression and Logistic Regression

```python
import numpy as np
```

## Linear Regression

```python
def linear_regression(X, w, b):
    return np.dot(w, X) + b
```

```python
def compute_cost_linear_regression(X, y, w, b, lambda_):
    m, n = X.shape
    cost = 0
    for i in range(m):
        f_wb_i = linear_regression(X[i], w, b)
        cost += (f_wb_i - y[i]) ** 2

    cost_reg = 0
    for j in range(n):
            cost_reg += w[j] ** 2
    cost_reg *= lambda_

    total_cost = (1 / (2 * m)) * (cost + cost_reg)
    return total_cost
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,6)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1]).reshape(-1,)-0.5
b_tmp = 0.5
lambda_tmp = 0.7
cost_tmp = compute_cost_linear_regression(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print("Regularized cost:", cost_tmp)
```

## Logistic Regression

```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g

def logistic_regression(X, w, b):
    z = np.dot(X, w) + b
    f_wb_x = sigmoid(z)
    return f_wb_x
```

```python
def compute_cost_logistic_regression(X, y, w, b, lambda_):
    m, n = X.shape
    cost = 0

    for i in range(m):
        f_wb_i = logistic_regression(X[i], w, b)
        cost += (y[i] * np.log(f_wb_i)) + (1 - y[i]) * (np.log(1 - f_wb_i))
    cost *= (-1 / m)

    cost_reg = 0
    for j in range(n):
        cost_reg += w[j] ** 2
    cost_reg *= (lambda_ / (2 * m))

    total_cost = cost + cost_reg
    return total_cost
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,6)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1]).reshape(-1,)-0.5
b_tmp = 0.5
lambda_tmp = 0.7
cost_tmp = compute_cost_logistic_regression(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print("Regularized cost:", cost_tmp)
```

## Gradient Computation

```python
def compute_gradient_linear_regression(X, y, w, b, lambda_):
    m, n = X.shape
    dj_dw = np.zeros((n, ))
    dj_db = 0
    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b
        err_i = f_wb_i - y[i]
        for j in range(n):
            dj_dw[j] += err_i * X[i, j]
        dj_db += err_i

    dj_dw /= m
    dj_db /= m

    for j in range(n):
        dj_dw[j] += (lambda_ / m) * w[j]

    return dj_dw, dj_db
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,3)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1])
b_tmp = 0.5
lambda_tmp = 0.7
dj_dw_tmp, dj_db_tmp =  compute_gradient_linear_regression(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print(f"dj_db: {dj_db_tmp}", )
print(f"Regularized dj_dw:\n {dj_dw_tmp.tolist()}", )
```

```python
def compute_gradient_logistic_regression(X, y, w, b, lambda_):
    m, n = X.shape
    dj_dw = np.zeros((n, ))
    dj_db = 0

    for i in range(m):
        f_wb_i = logistic_regression(X[i], w, b)
        err_i = f_wb_i - y[i]
        for j in range(n):
            dj_dw[j] += err_i * X[i, j]
        dj_db += err_i

    dj_dw /= m
    dj_db /= m

    for j in range(n):
        dj_dw[j] += (lambda_ / m) * w[j]

    return dj_dw, dj_db
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,3)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1])
b_tmp = 0.5
lambda_tmp = 0.7
dj_dw_tmp, dj_db_tmp =  compute_gradient_logistic_regression(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print(f"dj_db: {dj_db_tmp}", )
print(f"Regularized dj_dw:\n {dj_dw_tmp.tolist()}", )
```
